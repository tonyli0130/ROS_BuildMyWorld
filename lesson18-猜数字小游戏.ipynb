```python
import random

n = random.randint(1,100)#生成一个1-100的随机整数 
step = 0
high = 100#!!!!!!!!!!!!!!!!!!!!!!
low = 1#!!!!!!!!!!!!!!!!!!!!!!
print("Game start")

def get_number():
    guess = input("Please enter an integer from 1 to 100:")
    while True:
        if guess.isdigit():
            guess = int(guess)
            return guess
        else:
            guess = input("Please enter an integer from 1 to 100:")

guess = get_number()
            
while True:
    step += 1
    print("step",step)
    
    if guess == 0:
        print('quit')
        break
        
    if guess < n:
        print(guess,"is low")
        low = guess+1#!!!!!!!!!!!!!!!!!!!!!!
    elif guess > n:
        print(guess,"is high")
        high = guess-1#!!!!!!!!!!!!!!!!!!!!!!
    else:
        print("You win!")
        break
    print("You can try",low,"to",high)    #!!!!!!!!!!!!!!!!!!!!!!
    guess = get_number()
print("Game over")
```
